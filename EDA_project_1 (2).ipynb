```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
lowest_salary_female = grouped_data[grouped_data['Gender'] == 'f'].loc[grouped_data[grouped_data['Gender'] == 'f']['min'].idxmin()]
lowest_salary_male = grouped_data[grouped_data['Gender'] == 'm'].loc[grouped_data[grouped_data['Gender'] == 'm']['min'].idxmin()]
```

```python
print("\nLowest Salary for Female:")
print(lowest_salary_female)
print("\nLowest Salary for Male:")
print(lowest_salary_male)
```

```python

highest_salary_female = grouped_data[grouped_data['Gender'] == 'f'].loc[grouped_data[grouped_data['Gender'] == 'f']['min'].idxmin()]
highest_salary_male = grouped_data[grouped_data['Gender'] == 'm'].loc[grouped_data[grouped_data['Gender'] == 'm']['min'].idxmin()]
```

```python
print("\nHighest Salary for Female:")
print(highest_salary_female)
print("\nHighest Salary for Male:")
print(highest_salary_male)
```

```python
average_salary_designation = df.groupby('Designation')['Salary'].mean().reset_index()
print("\nAverage Salary for each Designation:")
print(average_salary_designation)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(highest_salary_female['Designation'], highest_salary_female['max'], color='blue', label='Female')
plt.bar(highest_salary_male['Designation'], highest_salary_male['max'], color='red', label='Male', alpha=0.7)
plt.title('Highest Salary by Designation and Gender')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(lowest_salary_female['Designation'], lowest_salary_female['min'], color='salmon', label='Female')
plt.bar(lowest_salary_male['Designation'], lowest_salary_male['min'], color='lightcoral', label='Male', alpha=0.7)
plt.title('Lowest Salary by Designation and Gender')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(average_salary_designation['Designation'], average_salary_designation['Salary'], marker='o', linestyle='-', color='blue')
plt.title('Average Salary by Designation')
plt.xlabel('Designation')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
